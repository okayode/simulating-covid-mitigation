```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def savefig(filename, crop = True):
    plt.savefig('{}.pdf'.format(filename))
    plt.savefig('{}.eps'.format(filename))
```

```python

resultsA = np.arange(0,30000,1)
results3 = np.loadtxt("DepthTrainA.txt") 
results4 = np.loadtxt("DepthtestA.txt") 
results5 = np.loadtxt("DepthTrainB.txt") 
results6 = np.loadtxt("DepthtestB.txt") 
results7 = np.loadtxt("DepthTrainC.txt") 
results8 = np.loadtxt("DepthtestC.txt") 
results9 = np.loadtxt("DepthTrainD.txt") 
results10 = np.loadtxt("DepthtestD.txt")

results3d = np.loadtxt("WidthTrainA.txt") 
results4d = np.loadtxt("WidthtestA.txt") 
results5d = np.loadtxt("WidthTrainB.txt") 
results6d = np.loadtxt("WidthtestB.txt") 
results7d = np.loadtxt("WidthTrainC.txt") 
results8d = np.loadtxt("WidthtestC.txt") 
#results9d = np.loadtxt("WidthTrainD.txt") 
#results10d = np.loadtxt("WidthtestD.txt")
```

```python
# All have this shape

print(resultsA.shape)

print(results3.shape)
print(results4.shape)
print(results5.shape)
print(results6.shape)
print(results7.shape)
print(results8.shape)
print(results9.shape)
print(results10.shape)

```

```python
fig, (ax3, ax4) = plt.subplots(nrows=1, ncols=2, sharex=True, figsize=(22, 9))

ax3.set_title('Fixed Width ',fontsize = 20)
ax3.semilogy(resultsA,np.mean(results3,0),'b')
ax3.fill_between(resultsA,
 np.mean(results3,0)-(1.96*np.std(results3,0)/np.sqrt(results3.shape[0])),
 np.mean(results3,0)+(1.96*np.std(results3,0)/np.sqrt(results3.shape[0])),
 alpha=0.5)
ax3.semilogy(resultsA,np.mean(results4,0),'b--')
ax3.fill_between(resultsA,
 np.mean(results4,0)-(1.96*np.std(results4,0)/np.sqrt(results4.shape[0])),
 np.mean(results4,0)+(1.96*np.std(results4,0)/np.sqrt(results4.shape[0])),
 alpha=0.5)
ax3.semilogy(resultsA,np.mean(results5,0),'g')
ax3.fill_between(resultsA,
 np.mean(results5,0)-(1.96*np.std(results5,0)/np.sqrt(results5.shape[0])),
 np.mean(results5,0)+(1.96*np.std(results5,0)/np.sqrt(results5.shape[0])),
 alpha=0.5)
ax3.semilogy(resultsA,np.mean(results6,0),'g--')
ax3.fill_between(resultsA,
 np.mean(results6,0)-(1.96*np.std(results6,0)/np.sqrt(results6.shape[0])),
 np.mean(results6,0)+(1.96*np.std(results6,0)/np.sqrt(results6.shape[0])),
 alpha=0.5)
ax3.semilogy(resultsA,np.mean(results7,0),'r')
ax3.fill_between(resultsA,
 np.mean(results7,0)-(1.96*np.std(results7,0)/np.sqrt(results7.shape[0])),
 np.mean(results7,0)+(1.96*np.std(results7,0)/np.sqrt(results7.shape[0])),
 alpha=0.5)
ax3.semilogy(resultsA,np.mean(results8,0),'r--')
ax3.fill_between(resultsA,
 np.mean(results8,0)-(1.96*np.std(results8,0)/np.sqrt(results8.shape[0])),
 np.mean(results8,0)+(1.96*np.std(results8,0)/np.sqrt(results8.shape[0])),
 alpha=0.5)
ax3.semilogy(resultsA,np.mean(results9,0),'k')
ax3.fill_between(resultsA,
 np.mean(results9,0)-(1.96*np.std(results9,0)/np.sqrt(results9.shape[0])),
 np.mean(results9,0)+(1.96*np.std(results9,0)/np.sqrt(results9.shape[0])),
 alpha=0.5)
ax3.semilogy(resultsA,np.mean(results10,0),'k--')
ax3.fill_between(resultsA,
 np.mean(results10,0)-(1.96*np.std(results10,0)/np.sqrt(results10.shape[0])),
 np.mean(results10,0)+(1.96*np.std(results10,0)/np.sqrt(results10.shape[0])),
 alpha=0.5)
ax3.set_xlabel('Epochs',fontsize = 20)
ax3.set_ylabel('MSE',fontsize = 20)
ax3.legend((r'Train 1 Hidden layer',r'Test 1 Hidden layer',\
            r'Train 2 Hidden layer',r'Test 2 Hidden layer',\
            r'Train 3 Hidden layer',r'Test 3 Hidden layer',\
            r'Train 4 Hidden layer',r'Test 4 Hidden layer',\
           ),loc='best',fontsize = 18)
ax3.tick_params(labelsize=20)


ax4.set_title('Fixed Depth ',fontsize = 20)
ax4.semilogy(resultsA,np.mean(results3d,0),'b')
ax4.fill_between(resultsA,
 np.mean(results3d,0)-(1.96*np.std(results3d,0)/np.sqrt(results3d.shape[0])),
 np.mean(results3d,0)+(1.96*np.std(results3d,0)/np.sqrt(results3d.shape[0])),
 alpha=0.5)
ax4.semilogy(resultsA,np.mean(results4d,0),'b--')
ax4.fill_between(resultsA,
 np.mean(results4d,0)-(1.96*np.std(results4d,0)/np.sqrt(results4d.shape[0])),
 np.mean(results4d,0)+(1.96*np.std(results4d,0)/np.sqrt(results4d.shape[0])),
 alpha=0.5)
ax4.semilogy(resultsA,np.mean(results5d,0),'g')
ax4.fill_between(resultsA,
 np.mean(results5d,0)-(1.96*np.std(results5d,0)/np.sqrt(results5d.shape[0])),
 np.mean(results5d,0)+(1.96*np.std(results5d,0)/np.sqrt(results5d.shape[0])),
 alpha=0.5)
ax4.semilogy(resultsA,np.mean(results6d,0),'g--')
ax4.fill_between(resultsA,
 np.mean(results6d,0)-(1.96*np.std(results6d,0)/np.sqrt(results6d.shape[0])),
 np.mean(results6d,0)+(1.96*np.std(results6d,0)/np.sqrt(results6d.shape[0])),
 alpha=0.5)
ax4.semilogy(resultsA,np.mean(results7d,0),'r')
ax4.fill_between(resultsA,
 np.mean(results7d,0)-(1.96*np.std(results7d,0)/np.sqrt(results7d.shape[0])),
 np.mean(results7d,0)+(1.96*np.std(results7d,0)/np.sqrt(results7d.shape[0])),
 alpha=0.5)
ax4.semilogy(resultsA,np.mean(results8d,0),'r--')
ax4.fill_between(resultsA,
 np.mean(results8d,0)-(1.96*np.std(results8d,0)/np.sqrt(results8d.shape[0])),
 np.mean(results8d,0)+(1.96*np.std(results8d,0)/np.sqrt(results8d.shape[0])),
 alpha=0.5)
ax4.set_xlabel('Epochs',fontsize = 20)
ax4.set_ylabel('MSE',fontsize = 20)
ax4.legend((r'Train 16 Neurons per layer',r'Test 16 Neurons per layer',\
            r'Train 32 Neurons per layer',r'Test 32 Neurons per layer',\
            r'Train 64 Neurons per layer',r'Test 64 Neurons per layer',\
           ),loc='best',fontsize = 18)
ax4.tick_params(labelsize=20)


savefig('./figures/italy_train_test_A')
```
